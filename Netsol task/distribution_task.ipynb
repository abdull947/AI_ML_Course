```python
import numpy as np
import pandas as pd

# 100 students (70% pass probability)
bernoulli_data = np.random.binomial(n=1, p=0.7, size=100)

df_bernoulli = pd.DataFrame({"Pass(1)/Fail(0)": bernoulli_data})
print(df_bernoulli.head(10))
```

```python
# Count the number of passes and fails
data = pd.DataFrame({"Pass": bernoulli_data})
print(data["Pass"].value_counts())
```

```python
# Visualize the distribution by histogram
import matplotlib.pyplot as plt

plt.figure()
plt.hist(data["Pass"], bins=2, edgecolor="black")
plt.title("Bernoulli Distribution (Pass/Fail)")
plt.xlabel("0 = Fail, 1 = Pass")
plt.ylabel("Number of Students")
plt.show()
```

**Bionomial Distribution**

```python
import numpy as np
import pandas as pd

np.random.seed(42)

# 100 days of data
# Each day: 20 customers, 60% chance of buying
binomial_data = pd.DataFrame({
    "Buyers_Per_Day": np.random.binomial(n=20, p=0.6, size=100)
})

print(binomial_data.head())
```

```python
print(binomial_data["Buyers_Per_Day"].describe())
```

```python
print(binomial_data["Buyers_Per_Day"].value_counts())
```

```python
#Visualize the distribution
import matplotlib.pyplot as plt
plt.figure()
plt.hist(binomial_data["Buyers_Per_Day"], bins=range(0, 21), edgecolor="black")
plt.title("Binomial Distribution (Buyers per Day)")
plt.xlabel("Number of Buyers")
plt.ylabel("Number of Days")
plt.show()
```

**Exponential Distribution**
Suppose bus stop pe average 10 minutes me 1 bus aati hai
Exponential distribution batata hai: next bus kitni der me aayegi

```python
#   Bus gaps (Exponential distribution)
import numpy as np
import pandas as pd

np.random.seed(42)

# Average bus every 10 minutes → λ = 1/10
lam = 1/10

# Simulate 100 bus gaps (in minutes)
bus_gaps = pd.DataFrame({
    "Time_Between_Buses": np.random.exponential(scale=1/lam, size=100)
})

print(bus_gaps.head())
```

```python
print(bus_gaps.describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.hist(bus_gaps["Time_Between_Buses"], bins=15, color='#ff6b6b', edgecolor='black')
plt.title("Exponential Distribution — Time Between Buses")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```

**T distribution**
suppose 8 apple and we measured the weight in gram 

```python
import numpy as np
import pandas as pd

# Small sample weights of apples
weights = np.array([120, 135, 128, 140, 125, 130, 138, 132])

t_data_fruits = pd.DataFrame({
    "Apple_Weight": weights
})

print(t_data_fruits)
```

```python
print(pd.DataFrame(weights).describe())
```

```python
from scipy import stats
import matplotlib.pyplot as plt

df = len(weights) - 1  # 7
x = np.linspace(min(weights)-5, max(weights)+5, 200)
t_pdf = stats.t.pdf(x, df, loc=np.mean(weights), scale=np.std(weights, ddof=1))

plt.figure(figsize=(7,5))
plt.plot(x, t_pdf, color='#ff6b6b', lw=2)
plt.title("t-Distribution — Apple Weights (Small Sample)")
plt.xlabel("Weight (grams)")
plt.ylabel("Probability Density")
plt.grid(True, alpha=0.3)
plt.show()
```
