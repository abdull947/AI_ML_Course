 Task 1 — Calculate Basic Statistics

```python
# Task 1 — Your code here
temps = [22, 25, 19, 30, 27, 21, 28, 24, 26, 23]

# 1. Mean
means=np.mean(temps)

# 2. Median
medians=np.median(temps)
print(f'Mean Score:{medians}')

# 3. Mode
modes=stats.mode(temps)
# 4. Print results
print(f'Mean Score:{means}')
print(f'Medians Score:{medians}')
print(f'Modes Score:{modes}')
```

 Task 2 — Measure the Spread

```python
# Task 2 — Your code here
temps = [22, 25, 19, 30, 27, 21, 28, 24, 26, 23]
mean = np.mean(temps)
# 1. Range
data_range = max(temps)-min(temps)
# 2. Variance
variance = np.var(temps)
print('\nStep-by-step manual calculation:')
sq_diffs = []
for x in temps:
    diff = x - mean
    sq   = diff ** 2
    sq_diffs.append(sq)
    print(f'{x:>7}  {diff:>12.1f}  {sq:>12.1f}')
print(f'\nSum of squared diffs: {sum(sq_diffs):.1f}')
print(f'Divided by n={len(temps)}: {sum(sq_diffs)/len(temps):.2f}  ✓')
# 3. Standard Deviation
Standard_deviation = np.std(temps)
# 4. Print summary sentence
print(f'Range:{data_range}')
print(f'Standard deviation:{Standard_deviation}')
```


⭐ Task 3 — Quartiles & Five-Number Summary

```python
# Task 4 — Your code here
salaries = [30000, 32000, 35000, 33000, 200000,
            31000, 34000, 36000, 29000, 32500]

# Q1, Q3, IQR
q1 = np.percentile(salaries,25)
q2 = np.percentile(salaries,50)
q3 = np.percentile(salaries,75)
IQR = q3 - q1
# Outlier fences
low_fence  = q1 - 1.5 * IQR
high_fence = q3 + 1.5 * IQR

outliers = [x for x in salaries if x < low_fence or x > high_fence]

# Detect and print outliers
print(f'Q1:{q1}')
print(f'Q1:{q2}')
print(f'Outliers found: {outliers if outliers else "None"}')
print(f'\nOutlier fences: {low_fence:.1f}  to  {high_fence:.1f}')
```

Task 5 — Z-Score Normalization

```python
# Task 5 — Your code here
heights = [150, 162, 170, 155, 180, 175, 168, 158, 172, 165]

# Compute mean and std
means = np.mean(heights)
Std = np.std(heights)
print(f'Mean: {means}  |  Std: {Std}\n')
# Print table header
print(f'  {"Height":>6}  {"Z-Score":>8}  {"Flag"}')
print('-' * 30)

# Calculate and print z-scores
# for h in heights:
for h in heights:
    z = (h - means)/Std
    if z > 1.5:
        label = 'unusual'
    else:
        label = ''
    print(f'{h:>7}  {z:>+9.8f}  {label}')
#     
#     print(...)
```

Task 6 — Explore Skewness

```python
# Task 6 — Your code here
from scipy import stats
right_skewed = [2, 3, 3, 4, 4, 4, 5, 5, 10, 15, 20]
left_skewed  = [1, 5, 10, 15, 16, 16, 17, 17, 18, 18, 19]

# For right_skewed:
for  name,data in [('Right Skewed',right_skewed)]:
     skew = stats.skew(data)
     mean = np.mean(data)
     med  = np.median(data)
     direction = 'positive →' if skew > 0.1 else 'negative'
     #for verifying the skew rule
     com = 'Mean > Median (right skew holds!)' if mean > med else 'Mean < Median (left skew holds!)'
     print(f'{name}\n  skewness={skew:+.2f}  direction: {direction} \n mean={mean:.1f}  median={med}\n {com}')
# For left_skewed:
for  name,data in [('Left Skewed',left_skewed)]:
     skew = stats.skew(data)
     mean = np.mean(data)
     med  = np.median(data)
     direction = 'positive →' if skew > 0.1 else 'negative'
    #for verifying the skew rule
     com = 'Mean > Median (right skew holds!)' if mean > med else 'Mean < Median (left skew holds!)'
     print(f'{name}\n  skewness={skew:+.2f}  direction: {direction} \n mean={mean:.1f}  median={med}\n {com}')

```

Task 7 — Correlation Analysis

```python
# Task 7 — Your code here
hours_sleep = [5, 6, 6, 7, 7, 8, 8, 9, 9, 10]
test_score  = [55, 60, 62, 68, 71, 75, 78, 82, 85, 88]

# 1. Pearson correlation
corr, pval = stats.pearsonr(hours_sleep, test_score)
print(f'Pearson r  = {corr:.3f}')
print(f'pval r= {corr:.3f}')
# 2. Strength classification
if corr >= 0.8:
    strength = 'STRONG positive correlation'
elif corr >= 0.5:
    strength = 'MODERATE'
else:
    strength = 'WEAK'

if corr > 0:
    direction = 'POSITIVE (as hours ↑, score ↑)'
else:
    direction = 'NEGATIVE (as hours ↑, score ↓)'
print(f'Strength : {strength}')
print(f'Direction: {direction}')
# 3. Significance check
print(f'Significant (p<0.05): {"YES ✅" if pval < 0.05 else "NO ❌"}')
# 4. (Bonus) Scatter plot
plt.figure(figsize=(6,4))
plt.scatter(hours_sleep, test_score, color='steelblue', s=80, zorder=5)
m, b = np.polyfit(hours_sleep,test_score, 1)
xline = np.array([min(hours_sleep), max(hours_sleep)])
plt.plot(xline, m*xline+b, color='red', linewidth=2, label=f'r = {corr:.2f}')
plt.xlabel('Hours Sleep')
plt.ylabel('Test Score')
plt.title('Hours Sleep vs Test Score')
plt.legend()
plt.tight_layout()
plt.show()
```
