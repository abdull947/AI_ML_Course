```python
#Quadratic equation
import cmath
a = float(input("Enter the coefficient a: "))
b = float(input("Enter the coefficient b: "))
c = float(input("Enter the coefficient c: "))
d = (b**2) - (4*a*c)
sqrt_d = cmath.sqrt(d)
root1 = (-b + sqrt_d) / (2*a)
root2 = (-b - sqrt_d) / (2*a)
print("The roots are:", root1, "and", root2)
```

```python
#use if else to check greater number
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
if num1 > num2:
    print(num1, "is greater than", num2)
elif num2 > num1:
    print(num2, "is greater than", num1)
else:
    print("Both numbers are equal")
    
```

```python
#use if else to check less number
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
if num1 < num2:
    print(num1, "is less than", num2)
elif num2 < num1:
    print(num2, "is less than", num1)
else:
    print("Both numbers are equal")
```

```python
# Prints all letters except 's' and 'i'
for st in "This is the python practice program and i am excited to do this task":
    if st == 's' or st == 'i':
       continue
    print(st,end = ' ')
```

```python
#use break
for st in "This is the python practice program and i am excited to do this task":
    if st == 'p':
      break
    print(st,end = ' ')
```
